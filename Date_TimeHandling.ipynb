```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#extract year
date['date_year'] = date['date'].dt.year
date.head()
```

```python
#extract month
date['date_month'] = date['date'].dt.month_name()
date.head()
```

```python
#extract day
date['date_day'] = date['date'].dt.day
date.head()
```

```python
#day of week
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
date['date_dow_name'] = date['date'].dt.day_name()
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#is Weekend ?
date['Date_weekend'] = np.where(date['date_dow_name'].isin(['Sunday','Saturday']),1,0)
date.drop(columns=['product_id','city_id','orders'])
```

```python
#extract week of year
date['date_week'] = date['date'].dt.weekday
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#extract quater
date['quarter'] = date['date'].dt.quarter
date.drop(columns=['product_id','city_id','orders']).head()
```
